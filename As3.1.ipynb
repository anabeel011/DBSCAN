```python
from sklearn.cluster import DBSCAN
import matplotlib.pyplot as plt
```

```python
data = [[2, 2, 8, 5, 7, 6, 1, 4],[10, 5, 4, 8, 5, 4, 2, 9]]
data2 = [[2,10], [2,5], [8,4], [5,8], [7,5], [6,4], [1,2], [4,9]]
plt.scatter(data[0], data[1])
plt.xlim([0, 10])
plt.title("Data before clustering")
```

```python
db = DBSCAN(eps = 2, min_samples = 2)
y = db.fit(data2)
labels = y.labels_
```

```python
color = {}
color[0] = 'r'
color[1] = 'g'
color[2] = 'b'
color[-1] = 'k'
cvec = [color[label] for label in labels] 
plt.scatter(data[0], data[1], c = cvec, cmap="rainbow")
plt.xlim([0,10])
plt.title('After clustering')
# black means the point is outlier
```
